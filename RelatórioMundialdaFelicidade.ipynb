# **1. Distribuição do Ladder Score**
Vamos começar com a distribuição do Ladder score, que é o índice de felicidade.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Ladder score'], kde=True, color='skyblue', bins=20)
plt.title('Distribuição do Ladder Score (Índice de Felicidade)', fontsize=14)
plt.xlabel('Ladder Score', fontsize=12)
plt.ylabel('Frequência', fontsize=12)
plt.show()
```

###*Propósito:*
Este gráfico mostra como os índices de felicidade (Ladder Score) estão distribuídos entre os países.

###*O que ele representa:*
Este gráfico mostra a felicidade dos países ao redor do mundo. As barras indicam quantos países têm um nível de felicidade em uma determinada faixa. A linha suavizada (KDE) mostra onde a maioria dos países está em termos de felicidade. Se as barras estão mais para a esquerda, significa que muitos países são menos felizes, enquanto se estão mais para a direita, significa que os países são mais felizes.

# **3. Relação entre Suporte Social e Ladder Score**


```python
import numpy as np


bins = np.linspace(df['Social support'].min(), df['Social support'].max(), 10)
labels = [f'{round(bins[i], 2)}-{round(bins[i+1], 2)}' for i in range(len(bins)-1)]
df['Social support bins'] = pd.cut(df['Social support'], bins=bins, labels=labels, include_lowest=True)

df_grouped = df.groupby('Social support bins', observed=False)['Ladder score'].mean()

plt.figure(figsize=(12, 6))
df_grouped.plot(kind='bar', color='orange')

plt.title('Relação entre Suporte Social e Ladder Score (Agrupado por Faixas)', fontsize=14)
plt.xlabel('Faixas de Social Support', fontsize=12)
plt.ylabel('Média do Ladder Score', fontsize=12)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

### *Propósito:*
Investigar se países com maior suporte social têm níveis mais altos de felicidade.

### *O que ele representa:*
Neste gráfico, o eixo X mostra o nível de suporte social em cada país, enquanto o eixo Y exibe o índice de felicidade (Ladder Score). Cada barra no gráfico representa um país.
